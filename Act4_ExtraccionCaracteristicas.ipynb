# **Agregar el archivo microretailer_mit_lift_lab.xlsx**

```python
df=pd.read_excel("microretailer_mit_lift_lab.xlsx")
df.head(5)
```

```python
#Selección de 20 variables categóricas
df_2= ['_updated_by','232_type_of_store','99_does_the_micro_retailer_exhibits_products_outside_','102_does_the_micro_retailer_sells_fresh_products_','184_store_devices','5_change_store_space_last_year','6_change_employees_average_salary_last_year','18_sales_records','145_number_direct_competitors','186_internet_connection', '49_inventory_records','210_sales_channels','189_payment_methods','214_customer_relationship_tools','193_sales_planning_tools','60_services','69_why_not_services','158_charact_organize_products_on_shelves','161_actions_stockouts','234_willingness_participate_workshop']
df2= df[df_2]
df2.head(5)
```

# **Aplicar métodos para reemplazar valores nulos del dataframe**

```python
#Identificación de nulos
total_nulos=df2.isnull().sum()
total_nulos
```

```python
#Limpieza nulos
df_retail=df2.fillna('Unknown')
```

```python
#Comprobación
total_nulos=df_retail.isnull().sum().sum()
total_nulos
```

# **Aplicar métodos para eliminar Outliers del dataframe**

```python
df_retail.info() #Al ser te tipo "Object" no se trataran outliers
```

# **Características a partir de aplicar un análisis univariado.**

Mostrar los resultados obtenidos ,mediante gráficas y tablas.



Columna '_updated_by' (Actualizado por)

```python
#Análisis univariado de una sola columna
tabla1=freq_tbl(df_retail['_updated_by'])
tabla1

#Piechart
Filtro1= tabla1.set_index('_updated_by')

# Filtrar las 20 categorías con las frecuencias más altas
top_20_categories = Filtro1.head(15)

# Crear el gráfico de pastel
top_20_categories['frequency'].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
plt.title('Gráfico de las personas con mayores actualizaciones')
plt.show()

```

Columna '232_type_of_store'

```python
tabla2= freq_tbl(df_retail['232_type_of_store'])
#Realizamos grafico de área pero solo filtramos los que tengan una frecuencia mayor a 1
filtro2= tabla2[tabla2['frequency']>1]
Filtro2_index= filtro2.set_index('232_type_of_store')

ax=Filtro2_index.plot(kind='area', figsize=(10,4),alpha = 0.5)
plt.show()
```

Columna '99_does_the_micro_retailer_exhibits_products_outside_'

```python
tabla3 = df_retail['99_does_the_micro_retailer_exhibits_products_outside_'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(tabla3.index, tabla3.values)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Los micro minoristas ,¿Exhiben productos afuera?')
plt.show()
```
